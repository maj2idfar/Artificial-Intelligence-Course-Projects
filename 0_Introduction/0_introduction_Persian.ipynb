```python
def with_vectorization():
    bachelors = df_original[df_original['education'] == 'Bachelors']
    
    return bachelors['hours-per-week'].mean()

print("Mean of bachelors' hours of work:", with_vectorization())
```

```python
def without_vectorization():
    count = 0
    sum = 0
    
    for index, row in df_original.iterrows():
        if row['education'] == 'Bachelors':
            sum += row['hours-per-week']
            count += 1

    return (sum / count)

print("Mean of bachelors' hours of work:", without_vectorization())
```

```python
print("Execution time of the way with vectorization:", timeit.timeit(with_vectorization, number=10))
print("Execution time of the way without vectorization:", timeit.timeit(without_vectorization, number=10))
```

```python
normalized.drop('salary', axis=1, inplace=True)

x = np.linspace(-5, 5, 100)

fig, axs = plt.pyplot.subplots(nrows=14, ncols=1, figsize=(14, 30))
fig.tight_layout(pad=3.0)

index = 0
for column in normalized:
    y1 = sp.stats.norm.pdf(x, lt_mean[column], lt_std[column])
    y2 = sp.stats.norm.pdf(x, mt_mean[column], mt_std[column])
    
    axs[index].plot(x, y1, label='<= 50', color='blue')
    axs[index].plot(x, y2, label='> 50', color='red')
    axs[index].set_title(column)

    index += 1
```

<div style="direction:rtl">
    حالا از باقی ستون ها صرف نظر میکنیم و ستون capital-gain را نرمال میکنیم.
</div>

```python
test_mean = test['capital-gain'].mean()
test_std = test['capital-gain'].std()
normalized_test = (test['capital-gain'] - test_mean) / test_std

normalized_test
```

<div style="direction:rtl">
    حالا به این صورت عمل میکنیم که ابتدا، دو تا محل تلاقی نمودارهای آبی و قرمز سوال 11 (برای capital-gain) را به دست می آوریم. در ادامه، همان طور که از نمودار هم مشخص است، اگر مقدار capital-gain بین این دو عدد باشد، به احتمال زیادی، درآمد فرد، کم تر از 50 هزار دلار است. در غیر این صورت احتمالا درآمد فرد بیشتر از 50 هزار دلار خواهد بود. (توجه به این که بیشتر داده هایی که در این بین واقع شده اند، آبی رنگ هستند). <br/>
    حالا این موضوع را برای تمام داده های فایل text.csv هم بررسی میکنیم و نتیجه را در یک دیتافریم جدید میریزیم.
</div>

```python
a = 1/(2*(lt_std['capital-gain']**2)) - 1/(2*(mt_std['capital-gain']**2))
b = mt_std['capital-gain']/(mt_std['capital-gain']**2) - lt_mean['capital-gain']/(lt_std['capital-gain']**2)
c = (lt_mean['capital-gain']**2)/(2*(lt_std['capital-gain']**2)) - (mt_std['capital-gain']**2)/(2*(mt_std['capital-gain']**2)) - np.log(mt_std['capital-gain']/lt_std['capital-gain'])

first_point, second_point = np.roots([a, b, c])

def predict(p):
    if first_point < p and p < second_point:
        return '<=50K'
    else:
        return '>50K'

result = pd.DataFrame()
result['salary'] = normalized_test.apply(predict)

result
```
